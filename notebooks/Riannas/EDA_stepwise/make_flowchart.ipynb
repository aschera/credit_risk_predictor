# Make a flowchart.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
data_processing_graph = nx.DiGraph()

# Add nodes (processing steps) to the graph
data_processing_graph.add_node("Data Collection. 35 columns. 800k rows.")
data_processing_graph.add_node("Data Cleaning")
data_processing_graph.add_node("Feature Engineering")
data_processing_graph.add_node("Model Training")
data_processing_graph.add_node("Model Evaluation")

# Define dependencies (edges) between nodes
data_processing_graph.add_edge("Data Collection. 35 columns. 800k rows.", "Data Cleaning")
data_processing_graph.add_edge("Data Cleaning", "Feature Engineering")
data_processing_graph.add_edge("Feature Engineering", "Model Training")
data_processing_graph.add_edge("Model Training", "Model Evaluation")
```

```python
# Create a larger figure
fig, ax = plt.subplots(figsize=(8, 8))

# Use nx.draw_networkx to draw the graph with labels
pos = nx.spring_layout(data_processing_graph, seed=42)
nx.draw_networkx(data_processing_graph, pos, with_labels=True, node_size=5000, node_color="skyblue", font_size=10, font_color="black", ax=ax)
ax.set_title("Data Processing Steps")
plt.axis("off")  # Turn off axis

# Display the plot
plt.show()
```
