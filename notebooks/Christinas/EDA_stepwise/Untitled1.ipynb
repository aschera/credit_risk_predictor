```python
import matplotlib.pyplot as plt

# Define the range of Fed's bond purchases
fed_bond_purchase = [i for i in range(0, 21, 2)]  # 0 to 20 billion in increments of 2 billion

# Calculate the potential increase in the money supply
money_multiplier = 5
potential_money_supply = [money_multiplier * purchase for purchase in fed_bond_purchase]

# Create the plot
plt.figure(figsize=(8, 6))
plt.bar(fed_bond_purchase, potential_money_supply, width=2, align='center')
plt.xlabel("Fed's Bond Purchase ($ billion)")
plt.ylabel("Potential Increase in Money Supply ($ billion)")
plt.title("Potential Increase in Money Supply vs. Fed's Bond Purchase")
plt.grid(True)

# Show the plot
plt.show()
```
