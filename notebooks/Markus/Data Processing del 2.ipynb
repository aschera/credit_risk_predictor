```python
# Calculate the total number of rows in the DataFrame
total_rows = len(df_encoded)

# Calculate the percentage of null values for each column
null_percentage = (df_encoded.isnull().sum() / total_rows) * 100

# Filter columns with more than 30% null values
columns_with_high_nulls = null_percentage[null_percentage > 30]

# Print columns with more than 30% null values and their respective null counts
print("Columns with more than 30% null values:")
for column, null_percent in columns_with_high_nulls.items():
    print(f"Column '{column}': {null_percent:.2f}% null values ({df_encoded[column].isnull().sum()} null values)")
```

```python
missing_percentage = (df_encoded.isnull().sum() / len(df_encoded) * 100)

columns_to_keep = missing_percentage[missing_percentage <= 41].index

new_data = df_encoded[columns_to_keep].copy()

df_cleaned = pd.DataFrame(new_data)

print(df_cleaned.info())
```

## NA handeling

```python
# census_tract is a geographical and therefore difficult to fill na with an average value. therefore any missing data will be dropped.

df_with_null_census = df_cleaned[df_cleaned['census_tract'].isnull()]

df_cleaned = df_cleaned.drop(df_with_null_census.index)
```

```python
# race, sex, age, ethnicity, income, and loan term are not values we want to standardise considering 
# that there are not a lot of missing data points in these columns, thus they will also be dropped.

df_cleaned = df_cleaned.drop(df_cleaned[(df_cleaned['applicant_ethnicity_1'].isna())
                    | (df_cleaned['co_applicant_ethnicity_1'].isna()) 
                    | (df_cleaned['applicant_race_1'].isna()) 
                    | (df_cleaned['co_applicant_race_1'].isna())
                    | (df_cleaned['applicant_sex'].isna())
                    | (df_cleaned['co_applicant_sex'].isna())
                    | (df_cleaned['applicant_age'].isna())
                    | (df_cleaned['co_applicant_age'].isna())
                    | (df_cleaned['income'].isna())
                    | (df_cleaned['loan_term'].isna())].index)
```

```python
df_cleaned.info()
```

```python
# Total units has one str value where one row is marked 5-24. 
# this row can be seen as an outlier and will be dropped.
print(df_cleaned['total_units'].value_counts())

df_cleaned = df_cleaned[df_cleaned['total_units']!= '5-24']

print(df_cleaned['total_units'].value_counts())
```

```python
df_cleaned.info()
```

```python
df_cleaned['origination_charges'] = pd.to_numeric(df_cleaned['origination_charges'], errors='coerce')
df_cleaned['interest_rate'] = pd.to_numeric(df_cleaned['interest_rate'], errors='coerce')
df_cleaned['combined_loan_to_value_ratio'] = pd.to_numeric(df_cleaned['combined_loan_to_value_ratio'], errors='coerce')
df_cleaned['property_value'] = pd.to_numeric(df_cleaned['property_value'], errors='coerce')
df_cleaned['total_loan_costs'] = pd.to_numeric(df_cleaned['total_loan_costs'], errors='coerce')
df_cleaned['loan_term'] = pd.to_numeric(df_cleaned['loan_term'], errors='coerce')
df_cleaned['total_units'] = pd.to_numeric(df_cleaned['total_units'], errors='coerce')
```
