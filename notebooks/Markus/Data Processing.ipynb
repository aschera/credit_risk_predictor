## Load Data 

```python
file = "C:/Users/marku/Documents/Programmering/Credit_Risk_Predictor/2019_public_lar_three_year_csv.csv"

condition = lambda x: x == 1
condition2 = lambda x: x == 2

df = pd.read_csv(file, converters={'loan_purpose': condition, 'business_or_commercial_purpose': condition2}, nrows = 1100)

```

## Dropping any non-key data fields

```python
# Columns to keep based on articles. 'loan_purpose' not included since it is already filtered

columns_to_keep = ['loan_type', 'loan_amount', 'action_taken', 'occupancy_type', 'census_tract', 'applicant_ethnicity_1', 
                   'co_applicant_ethnicity_1', 'applicant_race_1', 'applicant_race_2', 'co_applicant_race_1', 'co_applicant_race_2', 
                   'applicant_sex', 'co_applicant_sex', 'applicant_age', 'co_applicant_age', 'income', 'lien_status', 
                   'applicant_credit_score_type', 'co_applicant_credit_score_type', 'origination_charges', 
                   'discount_points', 'lender_credits', 'interest_rate', 'debt_to_income_ratio', 'combined_loan_to_value_ratio', 
                   'loan_term', 'property_value', 'manufactured_home_secured_property_type', 'total_units', 'aus_1', 'reverse_mortgage', 
                   'open_end_line_of_credit', 'manufactured_home_land_property_interest', 'total_loan_costs', 'total_points_and_fees', 
                   'prepayment_penalty_term', 'negative_amortization', 'interest_only_payment', 'balloon_payment', 
                   'other_nonamortizing_features', 'multifamily_affordable_units']
```

```python
# new variable with filtered columns

df_processed = df[columns_to_keep]
```

```python
df_encoded['co_applicant_age'].value_counts()
```

```python
df_encoded['debt_to_income_ratio'] = encoder.fit_transform(df_encoded['debt_to_income_ratio'])

df_encoded['debt_to_income_ratio'].value_counts()
```

## Changing Na values

```python
df_encoded.head()
```

```python
df_encoded['applicant_race_2'].fillna(0, inplace = True)
df_encoded['co_applicant_race_2'].fillna(0, inplace = True)
df_encoded['income'].fillna(0, inplace = True)
df_encoded['origination_charges'].fillna(0, inplace = True)
df_encoded['discount_points'].fillna(0, inplace = True)
df_encoded['lender_credits'].fillna(0, inplace = True)
df_encoded['interest_rate'].fillna(0, inplace = True)
df_encoded['combined_loan_to_value_ratio'].fillna(0, inplace = True)
df_encoded['loan_term'].fillna(0, inplace = True)
df_encoded['property_value'].fillna(0, inplace = True)
df_encoded['total_loan_costs'].fillna(0, inplace = True)
df_encoded['total_points_and_fees'].fillna(0, inplace = True)
df_encoded['prepayment_penalty_term'].fillna(0, inplace = True)
df_encoded['multifamily_affordable_units'].fillna(0, inplace = True)
```

```python
df_encoded['applicant_race_2'] = df_encoded['applicant_race_2'].astype(int)
df_encoded['applicant_race_2'].value_counts()
```
